```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv('./for_paper/csv/RQ1_2.csv', index_col=0)

print(df)

# choose colormap at seaborn
my_cmap = my_cmap = sns.color_palette("coolwarm", 6, as_cmap=True)

df.plot(kind='bar', figsize=(13,10), ylim=(0,1.19), colormap=my_cmap, fontsize=20)
plt.xticks(rotation=0)
plt.xlabel('Model', fontsize=20)
plt.ylabel('AUROC', fontsize=20)
plt.savefig('./RQ1_2.png')
```
